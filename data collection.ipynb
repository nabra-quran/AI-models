```python
import os
import requests
from bs4 import BeautifulSoup
```

# URLs

```python
url1 = "https://everyayah.com/data/warsh/warsh_Abdul_Basit_128kbps/"
url2 = "https://everyayah.com/data/warsh/warsh_ibrahim_aldosary_128kbps/"
url3 = "https://everyayah.com/data/warsh/warsh_yassin_al_jazaery_64kbps/"
```

# directories

```python
output_directory1 = "warsh_Abdul_Basit_mp3_files"
output_directory2 = "warsh_ibrahim_aldosary_mp3_files"
output_directory3 = "warsh_yassin_al_jazaery_mp3_files"
```

# Make directories

```python
os.makedirs(output_directory1, exist_ok = True)
os.makedirs(output_directory2, exist_ok = True)
os.makedirs(output_directory3, exist_ok = True)
```

# get responses

```python
response1 = requests.get(url1)
response2 = requests.get(url2)
response3 = requests.get(url3)
```

# start downloads

## reciter 1

```python
if response1.status_code == 200:
    soup = BeautifulSoup(response1.content, "html.parser")

    audio_links = soup.find_all("a", href = True)
    
    for link in audio_links:
        mp3_link = link["href"]
        
        if mp3_link.endswith(".mp3"):
            file_name = mp3_link.split("/")[-1]
            first_three_letters = file_name[:3]
            last_letters = file_name[3:]

            directory_path = os.path.join(output_directory1, first_three_letters) 

            os.makedirs(directory_path, exist_ok = True)
            
            mp3_response = requests.get(f'{url1}{file_name}')
            if mp3_response.status_code == 200:
                file_path = os.path.join(directory_path, last_letters)
                with open(file_path, "wb") as f:
                    f.write(mp3_response.content)
                    print("reciter 1 : ", file_name)
            else:
                print(f"Failed to download: {mp3_link}")
else:
    print("Failed to fetch the content. Status code:", response1.status_code)
```

## wav directories

```python
wav_output_directory1 = "warsh_Abdul_Basit_wav_files"
wav_output_directory2 = "warsh_ibrahim_aldosary_wav_files"
wav_output_directory3 = "warsh_yassin_al_jazaery_wav_files"
```

## Make wav directories

```python
os.makedirs(wav_output_directory1, exist_ok = True)
os.makedirs(wav_output_directory2, exist_ok = True)
os.makedirs(wav_output_directory3, exist_ok = True)
```

### imports

```python
from pydub import AudioSegment
```

```python
def mp3_to_wav(mp3_path, wav_path):
    audio = AudioSegment.from_mp3(mp3_path)
    audio.export(wav_path, format="wav")
```

```python
# Define the paths
mp3_directories = [ "warsh_Abdul_Basit_mp3_files",
                    "warsh_ibrahim_aldosary_mp3_files",
                    "warsh_yassin_al_jazaery_mp3_files"]
wav_directories = [ "warsh_Abdul_Basit_wav_files",
                    "warsh_ibrahim_aldosary_wav_files",
                    "warsh_yassin_al_jazaery_wav_files"]
```

## Start conversion

```python
for qaria_mp3_directory in mp3_directories:
    print(qaria_mp3_directory)
    for item in os.listdir(qaria_mp3_directory):
        surah_path = os.path.join(qaria_mp3_directory, item)

        folder_name = os.path.basename(surah_path)
        surah_wav_directory_path = os.path.join(wav_directories[mp3_directories.index(qaria_mp3_directory)], folder_name) 
        os.makedirs(surah_wav_directory_path, exist_ok = True)

        for verse in os.listdir(surah_path):
            verse_path = os.path.join(surah_path, verse)
            file_name = os.path.basename(verse_path)
            name = file_name[:3]

            wav_path = os.path.join(surah_wav_directory_path, f'{name}.wav')

            mp3_to_wav(verse_path, wav_path)
            print("Conversion complete: ", wav_path)
```
