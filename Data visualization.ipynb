```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
quran = pd.read_csv('ahkam_indexing/hisb_60_and_Al_fatihah_audio_with_transcript_and_MFCC_and_ahkam_indexing.csv')
```

```python
quran.columns
```

```python
quran.info()
```

```python
missing_values = quran.isnull().sum()
missing_values
```

```python
verse_distribution = quran.groupby(['recitor_en', 'sura_no']).size()
verse_distribution
```

```python
plt.figure(figsize=(3,5))
sns.set_style('darkgrid')
ax = sns.countplot(data=quran, y='recitor_en', palette='viridis')

for container in ax.containers:
    ax.bar_label(container, size=10, padding=2)

ax.set_title('Noumber of verses recited by each recitorer', fontweight='bold', fontsize=12)
ax.set_ylabel('Recitors', fontweight='bold')
ax.set_xlabel('Ayat', fontweight='bold')
ax.set_xticks([0, 50, 100, 150, 200, 250, 300])
        
plt.show()
```

```python
quran = pd.read_csv('warsh_transcript/abdul_basit_recitor_reciting_with_transcript.csv')
Al_Fatihah = quran[quran['sura_no'] == 1]
quran = quran[(quran['sura_no'] >= 87) & (quran['sura_no'] <= 114)]
df = pd.concat([Al_Fatihah, quran])
```

```python
plt.figure(figsize=(6,10))
sns.set_style('darkgrid')
ax = sns.countplot(data=df, y='sura_name_en', palette='husl')

for container in ax.containers:
    ax.bar_label(container, size=10, padding=2)

ax.set_title('Noumber of verses in each chapter', fontweight='bold', fontsize=12)
ax.set_ylabel('Chapters', fontweight='bold')
ax.set_xlabel('Verses', fontweight='bold')
ax.set_xticks([0, 5, 10, 15, 20, 25, 30, 35])
        
plt.show()
```
