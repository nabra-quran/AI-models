```python
import os
import pandas as pd
```

```python
ahkam_indexing = pd.read_excel('ahkam_indexing/Ahkam in ayat hisb 60 + 001_v2.xlsx')
```

```python
repeated_df = pd.concat([ahkam_indexing]*3, ignore_index=True)
columns_to_keep = ["madd_6_Lazim", "madd_246", "madd_6", "madd_2",
                    "Ikhfaa", "Idgham", "tafkhim", "qalqala", "imala"]
new_df = repeated_df[columns_to_keep]
```

```python
hisb60 = pd.read_csv('warsh_transcript/hisb_60_and_Al_fatihah_audio_with_transcript_and_MFCC.csv')
```

```python
special_characters = ['١', '٢', '٣', '٤', '٥', '٦', '٧', '٨', '٩', '٠', 'ۖ', '۞', '۩']
for char in special_characters:
    hisb60['aya_text'] = hisb60['aya_text'].str.replace(char, '')
hisb60['aya_text'] = hisb60['aya_text'].str.strip()
```
