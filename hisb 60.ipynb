```python
import os
import pandas as pd
```

```python
hisb_60_warsh_transcript_df = pd.read_excel('warsh_transcript/warshData_v2-1.xlsx', skiprows = 1)
hisb_60_warsh_transcript_df.reset_index(drop = True, inplace = True)
hisb_60_warsh_transcript_df = hisb_60_warsh_transcript_df.iloc[:-1]
Al_Fatihah = hisb_60_warsh_transcript_df[hisb_60_warsh_transcript_df['sura_no'] == 1]
hisb_60_warsh_transcript_df = hisb_60_warsh_transcript_df[(hisb_60_warsh_transcript_df['sura_no'] >= 87) & (hisb_60_warsh_transcript_df['sura_no'] <= 114)]

hisb_60_and_Al_Fatihah_warsh_transcript_df = pd.concat([Al_Fatihah, hisb_60_warsh_transcript_df])
```

```python
Al_Fatihah
```

```python
hisb_60_and_Al_Fatihah_warsh_transcript_df = hisb_60_and_Al_Fatihah_warsh_transcript_df[['sura_no',
                                            'sura_name_en',
                                            'sura_name_ar',
                                            'aya_no',
                                            'aya_text']]
hisb_60_and_Al_Fatihah_warsh_transcript_df.index = range(298)
hisb_60_and_Al_Fatihah_warsh_transcript_df
```

```python
directories = ["warsh_Abdul_Basit_wav_files", 
               "warsh_yassin_al_jazaery_wav_files_madani_hisb_60_and_Al_Fatihah", 
               "warsh_ibrahim_aldosary_wav_files_madani_hisb_60_and_Al_Fatihah"]
```

```python
abdul_basit_wav = []
yassin_al_jazaery_wav = []
ibrahim_aldosary_wav = []
wavs = [abdul_basit_wav, yassin_al_jazaery_wav, ibrahim_aldosary_wav]
for i in range(len(directories)) :
    if i == 0:
        subdir_name = "001"
        subdir_path = os.path.join(directories[i], subdir_name)
        if os.path.isdir(subdir_path):
            for filename in os.listdir(subdir_path):
                file_path = os.path.join(subdir_path, filename)
                file_path = file_path.replace("\\", "/")
                wavs[i].append(file_path)
        for j in range(87,115):
            subdir_name = str(j).zfill(3)
            subdir_path = os.path.join(directories[i], subdir_name)
            if os.path.isdir(subdir_path):
                for filename in os.listdir(subdir_path):
                    file_path = os.path.join(subdir_path, filename)
                    file_path = file_path.replace("\\", "/")
                    wavs[i].append(file_path)
    else : 
        for subdir_name in os.listdir(directories[i]):
            subdir_path = os.path.join(directories[i], subdir_name)
            if os.path.isdir(subdir_path):
                for filename in os.listdir(subdir_path):
                    file_path = os.path.join(subdir_path, filename)
                    file_path = file_path.replace("\\", "/")
                    wavs[i].append(file_path)
```

```python
abdul_basit_recitor = {
    'recitor_en': "Abdul Basit",
    'recitor_ar': "عبد الباسط",
    'recitor_reciting_aya_wav_path': wavs[0],
}

abdul_basit_recitor_df = pd.DataFrame(abdul_basit_recitor, index = None)
abdul_basit_recitor_df
```

```python
abdul_basit_hisb_60_and_Al_fatihah_audio_with_transcript = pd.concat([hisb_60_and_Al_Fatihah_warsh_transcript_df,
                                                                    abdul_basit_recitor_df], axis=1)
abdul_basit_hisb_60_and_Al_fatihah_audio_with_transcript
```

```python
yassin_al_jazaery_recitor = {
    'recitor_en': "Yassin Al Jazaery",
    'recitor_ar': "ياسين الجزائري",
    'recitor_reciting_aya_wav_path': wavs[1],
}

yassin_al_jazaery_recitor_df = pd.DataFrame(yassin_al_jazaery_recitor, index = None)
yassin_al_jazaery_recitor_df
```

```python
yassin_al_jazaery_hisb_60_and_Al_fatihah_audio_with_transcript = pd.concat([hisb_60_and_Al_Fatihah_warsh_transcript_df,
                                                                            yassin_al_jazaery_recitor_df], axis = 1)
yassin_al_jazaery_hisb_60_and_Al_fatihah_audio_with_transcript
```

```python
ibrahim_aldosary_recitor = {
    'recitor_en': "Ibrahim_Aldosary",
    'recitor_ar': "إبراهيم الدوسري",
    'recitor_reciting_aya_wav_path': wavs[2],
}

ibrahim_aldosary_recitor_df = pd.DataFrame(ibrahim_aldosary_recitor, index = None)
ibrahim_aldosary_recitor_df
```

```python
ibrahim_aldosary_hisb_60_and_Al_fatihah_audio_with_transcript = pd.concat([hisb_60_and_Al_Fatihah_warsh_transcript_df,
                                                                            ibrahim_aldosary_recitor_df], axis = 1)
ibrahim_aldosary_hisb_60_and_Al_fatihah_audio_with_transcript
```

```python
hisb_60_and_Al_fatihah_for_all_recitors_audio_path_and_transcript  = pd.concat([ibrahim_aldosary_hisb_60_and_Al_fatihah_audio_with_transcript,
                                                    yassin_al_jazaery_hisb_60_and_Al_fatihah_audio_with_transcript,
                                                    abdul_basit_hisb_60_and_Al_fatihah_audio_with_transcript])
hisb_60_and_Al_fatihah_for_all_recitors_audio_path_and_transcript 
```

```python
hisb_60_and_Al_fatihah_for_all_recitors_audio_path_and_transcript.to_csv('warsh_transcript/hisb_60_and_Al_fatihah_for_all_recitors_audio_path_and_transcript.csv', index = False)
```
