```python
import os
import pandas as pd
```

```python
def count_files_in_directory(directory):
    files = os.listdir(directory)
    file_count = sum(1 for file in files if os.path.isfile(os.path.join(directory, file)))
    return file_count
```

```python
directories = ["warsh_Abdul_Basit_mp3_files", "warsh_ibrahim_aldosary_mp3_files", "warsh_yassin_al_jazaery_mp3_files"]
directories_wav = ["warsh_Abdul_Basit_wav_files", "warsh_ibrahim_aldosary_wav_files", "warsh_yassin_al_jazaery_wav_files"]

dir_list = []
subdir_list = []
file_count_list = []
```

```python
subdir_counts = {}
for directory in directories:
    for root, dirs, files in os.walk(directory):
        sub_dir = root[len(directory):]
        file_count = count_files_in_directory(root)
        if sub_dir in subdir_counts:
            subdir_counts[sub_dir].append(file_count)
        else:
            subdir_counts[sub_dir] = [file_count]
recitor_sura_aya_count_df = pd.DataFrame(subdir_counts).transpose()
recitor_sura_aya_count_df.columns = directories

recitor_sura_aya_count_df['Same File Counts'] = recitor_sura_aya_count_df.apply(lambda row: 'Yes' if len(set(row)) == 1 else 'No', axis=1)

recitor_sura_aya_count_df
```

```python
recitor_sura_aya_count_filtered_df = recitor_sura_aya_count_df[recitor_sura_aya_count_df['Same File Counts'] == 'No']
pd.set_option('display.max_rows', None)
recitor_sura_aya_count_filtered_df
```
